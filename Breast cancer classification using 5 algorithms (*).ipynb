##Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

Import Dataset


```python
breast_cancer=pd.read_csv('data.csv')
breast_cancer
```

Data processing

```python
#Numbers of rows and columns in the dataset
breast_cancer.shape
```

```python
#Getting some information about our dataset
breast_cancer.info()
```

```python
#checking for missing columns in our dataset
breast_cancer.isna().sum()
```

from the previous code we get that "Unnamed column" have null values so we will drop it from our data 

```python
#droppinf 'unnamed' and 'id' columns
breast_cancer=breast_cancer.drop(["Unnamed: 32", 'id'], axis=1)
breast_cancer
```

```python
#conver "B","M" in Diagnosis column into 0,1
breast_cancer.diagnosis.replace(["M","B"],[1,0], inplace=True)
breast_cancer
```

```python
#statistical info about our dataset
breast_cancer.describe()
```

```python
#explore our target
breast_cancer.groupby("diagnosis").diagnosis.count()
```

```python
#Find the correlation among columns in dataset to remove the useless columns
corr=breast_cancer.corr(method ='pearson')
corr
```

```python
#checking the distribution of target variable
sns.countplot('diagnosis', data=breast_cancer)
plt.show()
```

```python




#plotting heatmap for correlation between features
plt.figure(figsize=(50,50))
plt.title('Correlation between features')
sns.heatmap(data=corr,annot=True,cmap="BuPu_r" ,annot_kws={'fontsize':20})
sns.set(font_scale=3) 
plt.show()
```

```python
#splitting the features and target
x=breast_cancer.drop(columns="diagnosis",axis=1)
y=breast_cancer['diagnosis']
```

```python
#splitting data into train&test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
##feature scalling
stc=StandardScaler()
x_train=stc.fit_transform(x_train)
x_test=stc.fit_transform(x_test)
```

1-Logisitic Regression

```python
#Trainig our model
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

2-Decision tree classifier

```python
dt_clf=DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)
```

```python
dt_clf.score(x_test,y_test)
```

```python
##plotting our features against the target to know the relation between them
pp = sns.pairplot(data=breast_cancer,height=5,
                  y_vars=['diagnosis'],
                  x_vars=['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean'])
```

```python
pp = sns.pairplot(data=breast_cancer,height=5,
                  y_vars=['diagnosis'],
                  x_vars=['radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst'])
```

there is a sigmoid relation between target and features not a linear relation 

To use KNN we need to know the number of neighbors 

```python
## The numbers of neighbours can be calculated by square root of the number of target values
import math
math.sqrt(len(y))
```

```python
##kNN model
knn_classifier=KNeighborsClassifier(n_neighbors=23,p=2,metric='euclidean')
knn_classifier.fit(x_train,y_train)

```

```python
knn_classifier.score(x_test,y_test)
```

```python
x_predict=knn_classifier.predict(x_train)
x_predict
```

```python
##to evaluate our model we need a conusion matrix
cm_knn=confusion_matrix(y_train,x_predict)
print(cm_knn)
```

```python
print(accuracy_score(y_train,x_predict))
```

4-SVM model

```python
svm=svm.SVC(kernel='linear',C=1)
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
svm_predict=svm.predict(x_train)
svm_predict
```

```python
cm_svm=confusion_matrix(y_train,svm_predict)
print(cm_svm)
```

```python

print(accuracy_score(y_train,svm_predict))
```
